````python
def age_score(Age, config_file):
    """ 
    age: risk factor with increasing age this subject DOES have
    config_file: The name of a configuration file to use for scoring these risk factors
    
    The function should return a score a points value for each age or age group 
    age= 75, points = 6
    """
    score = 0
    
    ### BEGIN SOLUTION
    
    config = json.load(open(config_file))
    age_rules = config.get("age")
    
    for rule in age_rules:
        if age >= rule.get('min') and age < rule.get('max'):
            score = rule.get('points')
    
    ### END SOLUTION

    return score
    
````
